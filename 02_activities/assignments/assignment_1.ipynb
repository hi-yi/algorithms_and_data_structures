#### Starter Code for Question 2

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(1)
root.left = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right = TreeNode(2)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)



def bt_path(root: TreeNode, path = [], result = []) -> list[list[int]]:
    # if there is no value for the node, exit
    if not root:
        return

    path.append(root.val) # append current node to path
    
    if not root.left and not root.right: # (base case) add completed path if the node is a leaf
        result.append(list(path))
    
    bt_path(root.left, path, result) # explore left first
    bt_path(root.right, path, result)
    path.pop() # go back if exploration is done

result = []
bt_path(root, [], result)
result
        
```

-   Paraphrase the problem in your own words


To return all root-to-leaf paths, we need to find every path from the root to the bottom of the binary tree. Since it's a binary tree, for each node, we must consider both the left and right child nodes.

Starting from the root (the top of the tree), we take one step at a time, exploring both the left and right branches. As we move down the tree, we keep track of the path. When we reach a leaf node (a node with no children), we have found a complete path from the root to that leaf, which we then store. This process continues until all root to leaf paths are found.


-   Code the solution to your assigned problem in Python (code chunk). Try to find the best time and space complexity solution!


```python
def bt_path(root: TreeNode, path = [], result = []) -> list[list[int]]:
    # If there is no value for the node, exit
    if not root:
        return

    # Append current node to path
    path.append(root.val) 
    
    # Add completed path if the node is a leaf
    if not root.left and not root.right: 
        result.append(list(path))
    
    # Explore left first
    bt_path(root.left, path, result) 
    bt_path(root.right, path, result)
    path.pop() # go back if exploration is done

result = []
bt_path(root, [], result)
result
```


-   Explain why your solution works


This function explores the tree from left to right. Looking at the code from the top, if there is no node, the function exits immediately. If the node exists, its value is added to the current path.

The third part of the function runs when the node is a leaf node. In this case, the completed path is added to the result list.

The last section is for exploration. The function first explores the left child, and if there is no left node, it moves to the right child. Once the function reaches a leaf node and completes the path, the last node is removed from the path to allow further exploration from the parent node.


-   Explain the problem’s time and space complexity


First, the algorithm explores all tree nodes, which takes O(N) time since each node is visited once.

At each leaf node, the current path from the root to that leaf is stored. In a balanced binary tree, the height is approximately O(log N), and each path takes O(log N) time per leaf.

Since there are approximately N/2 leaf nodes, the total time complexity is O(N + N/2 log N) => O(N log N)

Space complexity is O(N log N) as it needs O(log N) for each leaf.

The value of the current node is appended to the path. If the current node is a leaf, the current path is added to the result list. Finally, if there are children, these child nodes with the path are appended in path_list.
