```python
# trace/walkthrough partner's example

class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

tree_with_duplicates = TreeNode(1,
    TreeNode(2,
        TreeNode(4),
        TreeNode(2)  # Duplicate node
    ),
    TreeNode(3)
)

# return 2
# The value 2 appears in depth 1 and depth 2. (root.left, root.left.right)
```


-   Copy the solution your partner wrote. 


```python
def is_duplicate(root: TreeNode) -> int:
  def traverse(node, seen):
      if not node:
        return 0
      elif node.val in seen:
        return 1
      
      seen.add(node.val)
      return traverse(node.left, seen) or traverse(node.right, seen)
  
  return traverse(root, set())

print(is_duplicate(tree_with_duplicates)) 
```


-   Explain why their solution works in your own words.


> This code stores all values encountered during a depth-first traversal (DFS) from the left. If a value is already in the list, it means the same value has appeared in previously traversed nodes, so the function returns 1. If the traversal completes without finding any duplicate values, it returns 0. The code uses a recursive function to traverse all left and right children.


-   Explain the problem’s time and space complexity in your own words.


> Both time and space complexity are n as this code explores the entire tree only once.


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


> This code successfully returns 1 if there are duplicate values in the tree. However, the problem requires returning the duplicated value that is closest to the root, rather than just indicating whether duplicates exist.

> If using DFS as in this solution, all duplicated values along with their depths should be stored. After traversing all nodes, the closest duplicated value can be determined.

> Alternatively, BFS from root can be used, which does not require searching every node in the tree. Since BFS explores nodes level by level, it can return the first duplicated value immediately upon finding it.
